```python
folder = os.listdir('/Users/realgjl/Desktop/California/Original/SantaRosalia')

for file in folder:
    print(file)
    if file.endswith('jpg') or file.endswith('png'): # check the extension of a file
        image = cv2.imread('/Users/realgjl/Desktop/California/Original/SantaRosalia/' + file)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        kernel = np.array([[0, 0, -2, 0, 0],
                       [-2, -2, -2, -2, -2],
                       [-2, -2, 33, -2, -2],
                       [-2, -2, -2, -2, -2],
                       [0, 0, -2, 0, 0]])

        img = cv2.filter2D(image, -1, kernel)

        image_output = Image.fromarray(img)
        image_output.save("/Users/realgjl/Desktop/California/Sharpen/SantaRosalia/" + file)
    else:
        print("Pass the file with a name of " + file)
        pass
```
